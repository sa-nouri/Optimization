```python
import numpy as np
import pandas as pd


data = {
            "p1": ["I11", "I11", "I12", "I12"],
            "p2": ["I21", "I22", "I21", "I22"],
            "UL": [[2, 1], [2, 0], [0, 1], [0, 0]],
            "UR": [[0, 0], [0, 2], [2, 0], [2, 2]],
            "DL": [[0, 0], [0, 1], [1, 0], [1, 1]],
            "DR": [[1, 2], [1, 0], [0, 2], [0, 0]],
            "prob": [0.4, 0.1, 0.3, 0.2]
        
        }

df = pd.DataFrame(data)
df
```

```python
p1_all_str = ["UU", "UD", "DU", "DD"]
p2_strg = ["L", "R"]

result1 = []
for item in p1_all_str:
    s1, s2 = list(item)[0], list(item)[1]
    df1 = df[df["p2"] == "I21"]
    df2 = df[df["p2"] == "I22"]

    L1 = np.sum(np.vstack(df1[s1 + p2_strg[0]]), axis=0)[1] * df1["prob"].iloc[0]/df1.prob.sum()
    R1 = np.sum(np.vstack(df1[s1 + p2_strg[1]]), axis=0)[1] * df1["prob"].iloc[1]/df1.prob.sum()

    if L1 > R1:
        x = "L"
    elif L1 == R1:
        x = ["L", "R"]
    else:
        x = "R"

    L2 = np.sum(np.vstack(df2[s2 + p2_strg[0]]), axis=0)[1] * df2["prob"].iloc[0]/df2.prob.sum()
    R2 = np.sum(np.vstack(df2[s2 + p2_strg[1]]), axis=0)[1] * df2["prob"].iloc[1]/df2.prob.sum()

    if L2 > R2:
        y = "L"
    elif L2 == R2:
        y = ["L", "R"]
    else:
        y = "R"
    
    result1.append({
            "P1": item,
            "P2": x + y
    })
```

```python
p2_all_str = ["LL", "LR", "RL", "RR"]
p1_strg = ["U", "D"]

result2 = []

for item in p2_all_str:
    s1, s2 = list(item)[0], list(item)[1]
    df1 = df[df["p1"] == "I11"]
    df2 = df[df["p1"] == "I12"]

    U1 = np.sum(np.vstack(df1[p1_strg[0] + s1]), axis=0)[1] * df1["prob"].iloc[0]/df1.prob.sum()
    D1 = np.sum(np.vstack(df1[p1_strg[1] + s1]), axis=0)[1] * df1["prob"].iloc[1]/df1.prob.sum()

    if U1 > D1:
        x = "U"
    elif U1 == D1:
        x = ["U", "D"]
    else:
        x = "D"

    U2 = np.sum(np.vstack(df2[p1_strg[0] + s2]), axis=0)[1] * df2["prob"].iloc[0]/df2.prob.sum()
    D2 = np.sum(np.vstack(df2[p1_strg[1] + s2]), axis=0)[1] * df2["prob"].iloc[1]/df2.prob.sum()

    if U2 > D2:
        y = "U"
    elif U2 == D2:
        y = ["U", "D"]
    else:
        y = "D"
    
    result2.append({
            "P2": item,
            "P1": x + y
    })
```
